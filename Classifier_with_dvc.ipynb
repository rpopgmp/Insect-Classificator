```python
import os
params_content = """
data:
  train: train_data.pth
  test: test_data.pth

model:
  path: resnet18_insect_classifier.pth
  trained_path: resnet18_insect_classifier_trained.pth

training:
  lr: 0.001
  epochs: 10

metrics:
  confusion_matrix: confusion_matrix.npy
  accuracy_plot: class_accuracy_barplot.png
"""

with open(os.path.join('/content/drive/MyDrive/my_project', 'params.yaml'), 'w') as f:
    f.write(params_content)
```

```python
dvc_content = """
stages:
  prepare_data:
    cmd: python data_preparation.py
    deps:
      - /content/drive/MyDrive/data
    outs:
      - train_data.pth
      - test_data.pth

  create_model:
    cmd: python model_creation.py
    deps:
      - train_data.pth
      - test_data.pth
    outs:
      - resnet18_insect_classifier.pth

  train_model:
    cmd: python model_training.py
    deps:
      - resnet18_insect_classifier.pth
      - train_data.pth
    outs:
      - resnet18_insect_classifier_trained.pth
    params:
      - training.lr
      - training.epochs

  evaluate_model:
    cmd: python model_evaluation.py
    deps:
      - resnet18_insect_classifier_trained.pth
      - test_data.pth
    outs:
      - confusion_matrix.npy
      - class_accuracy_barplot.png
"""

# Сохранение файла dvc.yaml
with open(os.path.join('/content/drive/MyDrive/my_project', 'dvc.yaml'), 'w') as f:
    f.write(dvc_content)
```
